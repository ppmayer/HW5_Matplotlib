```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data
clinical_trial_data=pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data.head()
```

```python
# Read the Clinical Trial Data
mouse_drug_data=pd.read_csv(clinical_trial_data_to_load)
mouse_drug_data.head()
```

```python
# Combine the data into a single dataset
# Display the data table for preview
merged_data=pd.merge(mouse_drug_data,clinical_trial_data, on="Mouse ID",how="left")
merged_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
grouped_tumor_data=merged_data.groupby(['Drug','Timepoint'])
# Convert to DataFrame
grouped_tumor_data_for_mean=pd.DataFrame(grouped_tumor_data["Tumor Volume (mm3)"].mean()).reset_index()
grouped_tumor_data_for_mean.rename(columns={"Tumor Volume (mm3)":"Tumor Volume Mean (mm3)"},inplace=True)
# Preview DataFrame
grouped_tumor_data_for_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
grouped_tumor_data_for_SE=merged_data.groupby(['Drug','Timepoint'])[["Tumor Volume (mm3)"]].sem().reset_index()
grouped_tumor_data_for_SE.rename(columns={"Tumor Volume (mm3)":"Tumor Volume Standard Error (mm3)"},inplace=True)
# Preview DataFrame
grouped_tumor_data_for_SE.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
grouped_tumor_data_for_mean_rearranged=grouped_tumor_data_for_mean.pivot(index="Timepoint",columns="Drug")
grouped_tumor_data_for_mean_rearranged.columns = [col[1] for col in grouped_tumor_data_for_mean_rearranged.columns]
# Preview that Reformatting worked
grouped_tumor_data_for_mean_rearranged.head()
```

```python
# Create dataframe like above with standard errors per mean
grouped_tumor_data_for_SE_rearranged=grouped_tumor_data_for_SE.pivot(index="Timepoint",columns="Drug")
grouped_tumor_data_for_SE_rearranged.columns = [col[1] for col in grouped_tumor_data_for_SE_rearranged.columns]
# Preview that Reformatting worked
grouped_tumor_data_for_SE_rearranged.head()
```

```python
# Generate the Plot (with Error Bars)

plt.errorbar(grouped_tumor_data_for_mean_rearranged.index,grouped_tumor_data_for_mean_rearranged['Capomulin'], 
             yerr=grouped_tumor_data_for_SE_rearranged['Capomulin'], 
             fmt="o", color="r",linestyle='-',linewidth=0.5,marker='^')
plt.errorbar(grouped_tumor_data_for_mean_rearranged.index,grouped_tumor_data_for_mean_rearranged['Infubinol'], 
             yerr=grouped_tumor_data_for_SE_rearranged['Infubinol'], 
             fmt="o", color="b",linestyle='-',linewidth=0.5,marker='s')
plt.errorbar(grouped_tumor_data_for_mean_rearranged.index,grouped_tumor_data_for_mean_rearranged['Ketapril'], 
             yerr=grouped_tumor_data_for_SE_rearranged['Ketapril'], 
             fmt="o", color="g",linestyle='-',linewidth=0.5,marker='d')
plt.errorbar(grouped_tumor_data_for_mean_rearranged.index,grouped_tumor_data_for_mean_rearranged['Placebo'], 
             yerr=grouped_tumor_data_for_SE_rearranged['Placebo'], fmt="o", color="k",linestyle='-',linewidth=0.5)
plt.legend()
plt.grid()
plt.xlabel('Time (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Tumor Response to Treatment')
# Save the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Create dataframe like above with standard errors per mean
grouped_met_site_data_for_SE_rearranged=grouped_met_site_data_for_SE.pivot(index="Timepoint",columns="Drug")
grouped_met_site_data_for_SE_rearranged.columns = [col[1] for col in grouped_met_site_data_for_SE_rearranged.columns]
# Preview that Reformatting worked
grouped_met_site_data_for_SE_rearranged.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(grouped_met_site_data_for_mean_rearranged.index,grouped_met_site_data_for_mean_rearranged['Capomulin'], 
             yerr=grouped_met_site_data_for_SE_rearranged['Capomulin'], 
             fmt="o", color="r",linestyle='-',linewidth=0.5,marker='^')
plt.errorbar(grouped_met_site_data_for_mean_rearranged.index,grouped_met_site_data_for_mean_rearranged['Infubinol'], 
             yerr=grouped_met_site_data_for_SE_rearranged['Infubinol'], 
             fmt="o", color="b",linestyle='-',linewidth=0.5,marker='s')
plt.errorbar(grouped_met_site_data_for_mean_rearranged.index,grouped_met_site_data_for_mean_rearranged['Ketapril'], 
             yerr=grouped_met_site_data_for_SE_rearranged['Ketapril'], 
             fmt="o", color="g",linestyle='-',linewidth=0.5,marker='d')
plt.errorbar(grouped_met_site_data_for_mean_rearranged.index,grouped_met_site_data_for_mean_rearranged['Placebo'], 
             yerr=grouped_met_site_data_for_SE_rearranged['Placebo'], fmt="o", color="k",linestyle='-',linewidth=0.5)
plt.legend()
plt.grid()
plt.xlabel('Treatment Duration (days)')
plt.ylabel('Met Sites')
plt.title('Metastatic Spread During Treatment')
# Save the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
grouped_tumor_data_for_count=merged_data.groupby(['Drug','Timepoint'])[["Mouse ID"]].count().reset_index()
grouped_tumor_data_for_count.rename(columns={"Mouse ID":"Count of Mice ID"},inplace=True)

# Preview DataFrame
grouped_tumor_data_for_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
grouped_tumor_data_for_count_rearranged=grouped_tumor_data_for_count.pivot(index='Timepoint',columns='Drug')
grouped_tumor_data_for_count_rearranged.columns = [col[1] for col in grouped_tumor_data_for_count_rearranged.columns]
# Preview that Reformatting worked
grouped_tumor_data_for_count_rearranged.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(grouped_tumor_data_for_count_rearranged.index,
         grouped_tumor_data_for_count_rearranged['Capomulin']/grouped_tumor_data_for_count_rearranged['Capomulin'][0],
        color="r",linestyle='-',linewidth=0.5,marker='^')
plt.plot(grouped_tumor_data_for_count_rearranged.index,
         grouped_tumor_data_for_count_rearranged['Infubinol']/grouped_tumor_data_for_count_rearranged['Infubinol'][0],
        color="b",linestyle='-',linewidth=0.5,marker='s')
plt.plot(grouped_tumor_data_for_count_rearranged.index,
         grouped_tumor_data_for_count_rearranged['Ketapril']/grouped_tumor_data_for_count_rearranged['Ketapril'][0],
        color="g",linestyle='-',linewidth=0.5,marker='d')
plt.plot(grouped_tumor_data_for_count_rearranged.index,
         grouped_tumor_data_for_count_rearranged['Placebo']/grouped_tumor_data_for_count_rearranged['Placebo'][0],
        color="k",linestyle='-',linewidth=0.5,marker='o')

plt.legend()
plt.grid()
plt.xlabel('Time (days)')
plt.ylabel('Survival Rate (%)')
plt.title('Survival During Treatment')

# Show the Figure
plt.show()
```
